```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/car_crashes.csv"
car_crashes = pd.read_csv(url)
print(car_crashes.dtypes)
car_crashes_numeric = car_crashes.select_dtypes(include=['number'])
print(car_crashes_numeric.isnull().sum()) 
car_crashes_numeric = car_crashes_numeric.dropna()
correlation_matrix = car_crashes_numeric.corr()
plt.figure(figsize=(5, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='bwr', fmt=".2f", square=True)
plt.title('Correlation matrix of car crashes dataset')
plt.show()

print(car_crashes.head())
sns.set(style="darkgrid")
plt.figure(figsize=(5, 4))
sns.scatterplot(data=car_crashes, x='ins_premium', y='ins_losses', hue='alcohol', size='total', sizes=(10, 150), alpha=0.6)
plt.title('Insurance Losses vs. Premiums')
plt.xlabel('Insurance Premiums')
plt.ylabel('Insurance Losses')
plt.legend(title='Alcohol Impairment %')
plt.show()

plt.figure(figsize=(5, 4))
sns.boxplot(data=car_crashes, x='speeding', y='ins_losses')
plt.title('Insurance losses by speeding percentage')
plt.xlabel('Percentage of drivers speeding')
plt.ylabel('Insurance Losses')
plt.show()

sns.pairplot(car_crashes, vars=['total', 'speeding', 'alcohol', 'ins_premium', 'ins_losses'], hue='not_distracted')
plt.suptitle('Pairplot of key variables', y=1.02)
plt.show()
```
